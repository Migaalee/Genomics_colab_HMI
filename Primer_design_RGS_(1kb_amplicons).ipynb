Any package that you install with Conda will be installed into the directory /usr/local/lib/python3.6/site-packages so you will need to add this directory to sys.path in order for these packages to be available for import.

```python
import sys
_ = (sys.path
        .append("/usr/local/lib/python3.7/site-packages"))
```

### Running pipeline for 1kb amplicons

```python
%%shell
source activate primer
python /content/drive/MyDrive/NGS-PrimerPlex-master/NGS_primerplex.py \
    --threads 8 \
    -ref /content/drive/MyDrive/HMI/reference_wMelCs_b/wMelCs_b.fasta \
    -regions /content/drive/MyDrive/HMI/variants_wMelCs_b_1kb/wMelCs_b_1kb_targets.txt \
    --do-blast \
    -minampllen 700 \
    -maxampllen 1400 \
    -optampllen 1000 \
    -minprimerlen 17 \
    -maxprimerlen 27 \
    -optprimerlen 22 \
    -minprimermelt 55 \
    -maxprimermelt 60 \
    -optprimermelt 58 \
    -minprimergc 35 \
    -maxprimergc 65 \
    -optprimergc 40 \
    -primernum1 50 \
    -returnvariantsnum 2 \
    -run targets_X \
    -autoadjust 
```
